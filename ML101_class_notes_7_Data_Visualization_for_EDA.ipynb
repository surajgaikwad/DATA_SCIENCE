- Why data visualization is important.
- Try to understand it using **Anscombe's Quartet**



```python
# Anscombe's Quartet
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]

datasets = {
    'I': (x, y1),
    'II': (x, y2),
    'III': (x, y3),
    'IV': (x4, y4)
}


fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(6, 6),
                        gridspec_kw={'wspace': 0.08, 'hspace': 0.08})
axs[0, 0].set(xlim=(0, 20), ylim=(2, 14))
axs[0, 0].set(xticks=(0, 10, 20), yticks=(4, 8, 12))

for ax, (label, (x, y)) in zip(axs.flat, datasets.items()):
    ax.text(0.1, 0.9, label, fontsize=20, transform=ax.transAxes, va='top')
    ax.tick_params(direction='in', top=True, right=True)
    ax.plot(x, y, 'o')

    # linear regression
    p1, p0 = np.polyfit(x, y, deg=1)  # slope, intercept
    ax.axline(xy1=(0, p0), slope=p1, color='r', lw=2)

    # add text box for the statistics
    stats = (f'$\\mu$ = {np.mean(y):.2f}\n'
             f'$\\sigma$ = {np.std(y):.2f}\n'
             f'$r$ = {np.corrcoef(x, y)[0][1]:.2f}')
    bbox = dict(boxstyle='round', fc='blanchedalmond', ec='orange', alpha=0.5)
    ax.text(0.95, 0.07, stats, fontsize=9, bbox=bbox,
            transform=ax.transAxes, horizontalalignment='right')

plt.show()
```

**Explanation**: The above figure shows that the mean, standard deviation, and Pearson correlation value is the same. If we do not create figures, we will assume that they are equal. From the above figures, it is clear that they are different.

### Error bar
Error bars are used in charts and graphs to indicate the level of uncertainty or variability in the data being plotted. They provide a visual representation of the range of values that are likely to include the true value, based on a certain level of confidence.

```python
import plotly.graph_objs as go

trace = go.Bar(
    x=['A', 'B', 'C'],
    y=[4, 5, 6],
    error_y=dict(
        type='data',
        array=[0.1, 0.2, 0.3],
        visible=True
    )
)

data = [trace]

layout = go.Layout(
    title='Error Bar Example'
)

fig = go.Figure(data=data, layout=layout)
fig.show()

```

### Line plot
A line plot is used to see a graphical representation of data that uses a series of points connected by lines to show the values of a continuous variable over a period of time or across different categories. It helps visualize trends and patterns in the data, such as changes over time or differences between groups. It is also useful for identifying outliers or for comparing the values of multiple variables. Line plots can be used to display both time-series data and cross-sectional data.

### Scatter plot
A scatter plot is a graphical representation of data that uses dots to indicate the values of two continuous variables. It is used to show the relationship between the two variables, such as correlation or association. Scatter plots can be useful for identifying patterns and outliers, and for understanding the distribution of the data. They are widely used in statistics and data analysis to explore relationships between variables and to visualize the distribution of a dataset.

```python
fig = px.scatter(Gapminder_2007, x="gdpPercap", y="lifeExp", width=800, height=400,
            labels={"lifeExp": "Life Expectancy in year", "gdpPercap": "GDP per capita"})
fig.show()
```

### Heatmap 
- A heatmap is a graphical representation of data that uses color to indicate the values of two or more variables. It is used to display the relationship between the variables and to identify patterns and outliers in the data. It is useful for visualizing large and complex datasets.

```python
iris.corr()# By default it is a person correlation
# We only use numerical values, here we also have categorical values
# Has to follow pearson correlation assumptions, otherwise wrong
# species id is categorical in nature hence pearson correlation is not correct
# all assumptions have to be checked for pearson correlation
```

```python
import plotly.figure_factory as ff
iris_corr = iris.corr() # Generate correlation matrix for pearson correlation

x = list(iris_corr.columns)
y = list(iris_corr.index)
z = np.array(iris_corr)

fig = ff.create_annotated_heatmap(
    z,
    x = x,
    y = y ,
    annotation_text = np.around(z, decimals=2),
    hoverinfo='z',
    colorscale='Viridis'
    )
fig.show()
```

```python
iris.corr(method="spearman")# selecting the method as a spearman rank correlation
```

```python
import plotly.figure_factory as ff
iris_corr = iris.corr(method="spearman") # Generate correlation matrix for spearman correlation

x = list(iris_corr.columns)
y = list(iris_corr.index)
z = np.array(iris_corr)

fig = ff.create_annotated_heatmap(
    z,
    x = x,
    y = y ,
    annotation_text = np.around(z, decimals=2),
    hoverinfo='z',
    colorscale='Viridis'
    )
fig.show()
```

### Treemap
A treemap is a graphical representation of data that uses nested rectangles of different sizes and colors to indicate the hierarchical structure and relative proportions of a dataset. It is used to display hierarchical data such as file directories, and allows to visualize the proportion of different categories within a dataset.

```python
fig = px.treemap(Gapminder_2007, path=[px.Constant("world"), 'continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'],
                  color_continuous_scale='RdBu',
                  color_continuous_midpoint=np.average(Gapminder_2007['lifeExp'], weights=Gapminder_2007['pop']))
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```
