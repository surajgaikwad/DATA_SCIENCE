## **Common unsupervised learning approaches**

Clustering, association, and dimensionality reduction are the three basic tasks that unsupervised learning models are used for.

  image.png

### **Clustering**

* Unlabeled data are grouped using the data mining technique of clustering according to their similarities or differences.
* Clustering is used to process raw data, creating groups of unclassified data by specific pattern in the information.

  image.png

* Clustering techniques are categorized into exclusive, overlapping, hierarchical and probabilistic.

### Exclusive clustering

* A data point may only be included in one cluster according to the grouping method known as exclusive clustering. Another name for this is "hard" clustering.
* Exclusive clustering is exemplified by the K-means algorithm.

### Overlapping clustering

* Data points can be members of many clusters with varying degrees of membership under overlapping clusters, which is different from exclusive clustering.
* Overlapping clustering is demonstrated by "soft" or fuzzy k-means clustering.

### Hierarchical clustering

* An unsupervised clustering approach known as hierarchical clustering, sometimes referred to as hierarchical cluster analysis (HCA), can be classified as either agglomerative or divisive. 
* A "bottoms-up technique" is referred known as agglomerative clustering. Its data points are initially separated into distinct groupings, and once one cluster has been obtained, they are progressively combined based on similarity.
* The antithesis of agglomerative clustering, referred to as divisive clustering, operates from the top down. In this instance, divisions between data points inside a single data cluster are made.

 

### Probabilistic clustering

* It helps us to find the best density estimation or "soft" clustering problems.
* The likelihood method is used to cluster the data points which belong to a particular distribution.
* The Gaussian Mixture Model (GMM), one of the most used probabilistic clustering techniques, was developed in the 1960s.

image.png

* Data points are divided into K groups using the K-means clustering technique, which determines the number of clusters depending on the distance from the centroid of each group.
* The data points that fall into the same category are those that are closest to a certain centroid.
* Smaller K values indicate bigger groupings and less granularity, while greater K values indicate smaller groupings and higher granularity.
* It is an iterative approach that separates the unlabeled dataset into k distinct clusters, each of which contains just one dataset and shares a set of characteristics.
* It gives us the ability to divide the data into various groups and provides a practical method for automatically identifying the groups in the unlabeled dataset without the need for any training.
* Each cluster has a centroid assigned to it because the algorithm is centroid-based. This algorithm's primary goal is to reduce the total distances between each data point and its corresponding clusters.
* Market segmentation, document clustering, image segmentation, and image compression all frequently use K-means clustering.

## How does it work?

The algorithm starts with an unlabeled dataset as its input, separates it into k clusters, and then continues the procedure until it runs out of clusters to use. In this algorithm, the value of k should be predetermined.

The k-means clustering performs two tasks:

* Uses an iterative technique to choose the best value for K centre points or centroids.
* Each data point is matched with the nearest k-center. A cluster is formed by the data points that are close to a specific k-center.

As a result, each cluster is distinct from the others and contains datapoints with some commonality.

image.png

The steps in the K-means algorithm
- Step 1: To determine the number of clusters, choose K.
- Step 2: Choose K locations or centroids at random.
- Step 3: Assign each data point to its nearest centroid, which will create the K clusters that have been predetermined.
- Step 4: Determine the variance and relocate each cluster's centroid.
- Step 5: Re-assign each data point to the new centroid of each cluster by repeating the third step.
- Step 6: Go to step 4 if there is a reassignment; otherwise, move to FINISH.
- Step 7: The finished model.

## How to choose the value of "K number of clusters" in K-means Clustering?

How should "K number of clusters" be set in K-means Clustering?
The incredibly powerful clusters that the K-means clustering algorithm generates are what make it so effective. It can be challenging to choose the right amount of clusters, though. Although there are numerous methods for figuring out the ideal number of clusters, in this article we concentrate on the most effective one. The steps are explained below:

**Elbow Method**
The Elbow approach is one of the most popular methods for figuring out the optimal number of clusters. This technique makes use of the WCSS value principle. The acronym "WCSS," which stands for Within Cluster Sum of Squares, is used to refer to "total variances inside a cluster." The value of WCSS can be calculated using the formula below (for 3 clusters):

image.png

In the WCSS formula mentioned above,

**Pi in Cluster1 distance(Pi C1)^2**: It is the sum of the square of the distances between each data point and its centroid within a cluster1 and the same for the other two terms.

The distance between the data points and the centroid can be determined using techniques like the Manhattan distance or the Euclidean distance.

The following steps are followed by the elbow approach to determine the best cluster value:

    * It performs K-means clustering for different K values on a given dataset (ranges from 1-10).
    * Determine the WCSS score for each K value.
    * draws a curve connecting the estimated WCSS values with the K-cluster number. The greatest K value is assigned to a bend's sharp tip or a plot point that resembles an arm.



The graph shows a sharp bend that resembles an elbow, hence the term "elbow technique." The graph produced by the elbow method looks like the example below:

image.png

**Note**: We can choose a cluster size that is appropriate for the given number of data points. If we choose an equal number of clusters to the number of data points, the plot's endpoint will be achieved, and the WCSS value will be zero.

## Python Implementation of K-means Clustering Algorithm

### Step-1: Data pre-processing Step

```python
# importing libraries    
import numpy as nm    
import matplotlib.pyplot as mtp    
import pandas as pd    
```

```python
# Importing the dataset  
dataset = pd.read_csv('Mall_Customers.csv')  
```

```python
dataset.head()
```

From the above dataset, we need to find some patterns in it.

**Extracting Independent Variables** 

Although this is a clustering problem, we don't require any dependent variables for the data pre-processing stage because we don't know what to find. So, we will merely add a line of code for the features matrix.

```python
x = dataset.iloc[:, [3, 4]].values
```

We are simply extracting the third and fourth features, as can be seen. It's because some features, like customer id, are not necessary when creating a 2d plot to represent the model.

### Step-2: Finding the optimal number of clusters using the elbow method

The second stage will involve determining the ideal number of clusters to use for our clustering issue. As was said previously, we will employ the elbow approach in this instance.

As is well known, the elbow approach plots the WCSS values on the Y-axis and the number of clusters on the X-axis to create the plot using the WCSS idea. Hence, for various k values between 1 and 10, we will calculate the value for WCSS. The code is as follows:

```python
#finding optimal number of clusters using the elbow method  
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#Using for loop for iterations from 1 to 10.  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)  
mtp.plot(range(1, 11), wcss_list)  
mtp.title('The Elobw Method Graph')  
mtp.xlabel('Number of clusters(k)')  
mtp.ylabel('wcss_list')  
mtp.show()  
```

To create the clusters, we used the KMeans class of the sklearn.cluster library.

In order to establish an empty list that will hold the value of wcss computed for various values of k ranging from 1 to 10, we constructed the wcss list variable.

Next, as Python's for loop excludes the outbound limit, we initialised the for loop to iterate on a different value of k between 1 and 10; it is taken as 11 to include the 10th value. We have presented the relationship between the number of clusters and WCSS after fitting the model to a matrix of features.

From the above plot, we can see the elbow point is at 5. So the number of clusters here will be 5.

```python
wcss_list
```

### Step- 3: Training the K-means algorithm on the training dataset

We can now train the model on the dataset because we know how many clusters there are.

The two lines of code we used in the section above will be used again to train the model, but this time we will use 5 instead of I because we know that 5 clusters must be produced. 

```python
#training the K-means model on a dataset  
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x)  
```

When constructing an object of the KMeans class, the first line is the same as above.

The dependent variable y predict was established in the second line of code to train the model.

We will obtain the y predict variable by running the code in the aforementioned lines. In the Spyder IDE, we can inspect it using the variable explorer option. Now that we have our original dataset and y predict values, we can compare them.

```python
y_predict
```

As the index starts at 0, we can now relate that CustomerID 1 belongs to cluster 3, CustomerID 2 belongs to cluster 4, and so on.

Step-4: Visualizing the Clusters

The clusters must be visualised as the final phase. Since our model consists of 5 clusters, we will each visualise a cluster separately.

To visualize the clusters will use scatter plot using mtp.scatter() function of matplotlib.

```python
#visulaizing the clusters  
mtp.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
mtp.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
mtp.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
mtp.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster  
mtp.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') #for fifth cluster  
mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')   
mtp.title('Clusters of customers')  
mtp.xlabel('Annual Income (k$)')  
mtp.ylabel('Spending Score (1-100)')  
mtp.legend()  
mtp.show()  
```
