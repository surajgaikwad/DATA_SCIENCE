# Missing values
- Missing values are a most necessary part of raw data. Technically, it should not be, but we have to deal with missing values in many scenarios.
- When dealing with missing values, we must have to be cautious. One approach is only sometimes the correct one. Even in the same data set, you may have to deal with different variables in different way.
- The different domains had different approaches to dealing with missing values. For example, in genomics research, if more than 1% of data is missing, we prefer to remove them. In other fields, we still like to replace them until more than 30% to 40% of data is missing.

## Replace missing values 

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguin_raw = pd.read_excel('ML101 Dataset_2 penguin_manipulated_data_set.xlsx') # reading dataset
penguin = penguin_raw.copy() # creating copy of original file
```

```python
penguin.isnull() # checking if null values are present or absent
```

```python
penguin[penguin["Culmen Depth (mm)"].isnull()]
```

```python
penguin
```

Here, true suggests missing values are present and false represent actual values.

```python
# dropping unnecessary columns
penguin = penguin.drop(['studyName', 'Sample Number', 'Stage', 'Region', 'Date Egg', 'Individual ID', 'Delta 15 N (o/oo)', 'Delta 13 C (o/oo)', 'Comments'], axis = 1)
```

```python
penguin.isnull().sum() # Adding all trues as true means one and false means zero. 
```

```python
# renaming columns
penguin.rename(columns = {'Clutch Completion':'Clutch_Completion', 'Culmen Length (mm)':'Culmen_Length', 
                         'Culmen Depth (mm)':'Culmen_Depth', 'Flipper Length (mm)':'Flipper_Length', 'Body Mass (g)':'Body_Mass'}, inplace = True)
```

```python
penguin.info() # checking if non-null count and data types. 
```

- Body mass is numeric values but python is reading it as object. Hence, possibility of special characters.

```python
BMI_unique = penguin.Body_Mass.unique() # Here unique function to find unique values
BMI_unique
```

- There is '.' present in the column, hence python is reading it as object.

```python
# Replacing . with NaN
penguin = penguin.replace('\.+', np.nan, regex=True)
penguin.head()
```

```python
penguin['Body_Mass'] = pd.to_numeric(penguin['Body_Mass'], errors='coerce') # converting Body_mass into numeric
```

```python
penguin.info() # checking info again
```

```python
penguin.describe(include = 'all') # summarizing all variables
```

```python
sns.histplot(data = penguin['Body_Mass'], kde = True) # creating histogram with desnity plot
plt.axvline(x=penguin.Body_Mass.mean(),color='red',alpha=0.5,label='Mean') # adding vartical line as mean
plt.axvline(x=penguin.Body_Mass.median(),c='blue',ls='--',alpha=0.5,label='Median') # adding vartical line as median
plt.legend() # adding legend
```

- Histogram suggests there are many values entered as 0 for BMI. It seems missing values may be entered as 0. Hence, these values should be replaced as NaN. 

```python
penguin.Body_Mass[penguin.Body_Mass == 0] # Filtering zero in Body_mass column
```

```python
penguin.Body_Mass[penguin.Body_Mass == 0] = np.nan # replacing 0 with np.nan
penguin.Body_Mass[np.isnan(penguin.Body_Mass)] # checking nan 
```

- Missing values can be of multiple types


**1- Missing completely at random (MCAR)**: "Missingness has no relationship between any values, observed or missing."

**2- Missing at Ranodm (MAR)**: "There is a systematic relationship between missingness and other observed data, but not
the missing data".

**3- Missing not at Random (MNAR)**: "There is a relationship between missingness and its values, missing or non-missing."

### Creating dummy variables

```python
## Converting males to 0 and females to 1
penguin['Sex'] = penguin.Sex.map({'MALE': 0, 'FEMALE':1})

## Converting clutch completion no = 0  and yes = 1
penguin['Clutch_Completion'] = penguin.Clutch_Completion.map({'No': 0, 'Yes':1})

## Converting species Adelie = 0 chinstrap = 1, and Gentoo = 2
penguin['Species'] = penguin.Species.map({'Adelie Penguin (Pygoscelis adeliae)': 0, 'Chinstrap penguin (Pygoscelis antarctica)':1,
                                                   'Gentoo penguin (Pygoscelis papua)':2})

## Converting Island Biscoe = 0 Dream = 1, and Torgersen = 2
penguin['Island'] = penguin.Island.map({'Biscoe': 0, 'Dream':1,
                                                   'Torgersen':2})
```

### Basic imputation techniques

#### Mean imputation

```python
from sklearn.impute import SimpleImputer # importing simple imputer
penguin_mean = penguin.copy(deep=True) # creating a copy for further evaluation
mean_imputer = SimpleImputer(strategy='mean') # using mean to impute the value
penguin_mean.iloc[:, :] = mean_imputer.fit_transform(penguin_mean) #  both fit a model to the data and then transform the data using that model
```

#### Median imputation

```python
penguin_median = penguin.copy(deep=True) # creating a copy for further evaluation
median_imputer = SimpleImputer(strategy='median') # using median to impute the value
penguin_median.iloc[:, :] = median_imputer.fit_transform(penguin_median)
```

#### Mode imputation

```python
penguin_mode = penguin.copy(deep=True) # creating a copy for further evaluation
mode_imputer = SimpleImputer(strategy='most_frequent') # using most frequent to impute the value
penguin_mode.iloc[:, :] = mode_imputer.fit_transform(penguin_mode)
```

#### Constant imputation

```python
penguin_constant = penguin.copy(deep=True) # creating a copy for further evaluation
constant_imputer = SimpleImputer(strategy='constant', fill_value = 0) # using constant to impute the value
penguin_constant.iloc[:, :] = constant_imputer.fit_transform(penguin_constant)
```

### Advance Imputation techniques

```python
from fancyimpute import KNN # importing KNN library
knn_imputer = KNN() # calling KNN function
penguin_knn = penguin.copy(deep=True) # creating copy
penguin_knn.iloc[:, :] = knn_imputer.fit_transform(penguin_knn)
```

```python
penguin_knn.head() # observing first five rows
```

- Provides numerical values, even for categorical data.
- For binary variable, we can treat anything < 0.5 as 0 and 0.5 >= as 1.
- This is just an example. **We never fill missing values for target variable.** We prefer to remove them.
- It is good to exclude y variables to avoid data leakage during advance imputation techniques.


**2- Multiple Imputation by chained equation**

- Perform multiple regressions over random sample of the data

- Take average of the multiple regression values

- Impute the missing feature value for the data point

```python
from fancyimpute import IterativeImputer # importing iterative imputers
MICE_imputer = IterativeImputer() # calling iterativeIMputer
penguin_MICE = penguin.copy(deep=True) # creating a copy of penguin
penguin_MICE.iloc[:, :] = MICE_imputer.fit_transform(penguin_MICE)
```

```python
penguin_MICE.head() # observing first five rows
```

- Provides numerical values, even for categorical data.
- For binary variable, we can treat anything < 0.5 as 0 and 0.5 >= as 1.

### Visualzing Imputation 

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 10)) # creating 2 rows and 3 columns
nullity = penguin['Culmen_Length'].isnull() + penguin['Culmen_Depth'].isnull() # creating null columns between culmen length and culmen depth
imputations = {'Mean Imputation': penguin_mean, # creating a python dictionary
               'Median Imputation': penguin_median,
               'Most Frequent Imputation': penguin_mode,
               'Constant Impu tation': penguin_constant,
               'KNN Imputation': penguin_knn,
              'MICE Imputation': penguin_MICE}

for ax, df_key in zip(axes.flatten(), imputations): # a for loop to iterate over the subplots and the imputed data
    imputations[df_key].plot(x='Culmen_Length', y='Culmen_Depth', kind='scatter',
                             alpha=0.5, c=nullity, cmap='rainbow', ax=ax,
                             colorbar=False, title=df_key)
```

- Visualization clearly suggest KNN imputation as a better approach to deal with missing values. However, this example is only applicable here as there is clearly a relationship in missingness between culmen length and culmen depth.

- For different variables a different approach may be better for missing value.
- For example, we can create box plot between missing value vs non missing values for numerical variable
- Bar plot for categorical variable.

- Overall justification is must when we are doing anything.
